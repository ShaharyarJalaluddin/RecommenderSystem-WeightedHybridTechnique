```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies.head()
```

```python
print("credits: ", credits.shape)
print("movies: ", movies.shape)
```

```python
credits_column_rename = credits.rename(index=str, columns={"movie_id":"id"})
movie_merge = movies.merge(credits_column_rename, on='id')
movie_merge.head()
```

```python
movie_merge.columns
```

```python
movies_cleaned = movie_merge.drop(columns=['homepage','title_x', 'title_y', 'status','production_countries'])
movies_cleaned.head()
```

```python
movies_cleaned.info()
```

```python
# calculate all the components based on the formula
v = movies_cleaned['vote_count']
R = movies_cleaned['vote_average']
C = movies_cleaned['vote_average'].mean()
m = movies_cleaned['vote_count'].quantile(0.70)
```

```python
movies_cleaned['Weighted_average'] = ((R*v) + (C*m)) / (v+m)
```

```python
movies_cleaned.head()
```

```python
movies_ranking = movies_cleaned.sort_values('Weighted_average', ascending=False)
movies_ranking[['original_title', 'vote_count','vote_average', 'Weighted_average', 'popularity']].head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

weight_average = movies_ranking.sort_values('Weighted_average', ascending = False)
plt.figure(figsize=(12,6))
axis1 = sns.barplot(x=weight_average['Weighted_average'].head(10), y = weight_average['original_title'].head(10))
plt.xlim(4, 10)
plt.title('Best Movies by avg votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.jpeg')
```

```python
popularity = movies_ranking.sort_values('popularity', ascending=False)
plt.figure(figsize=(12,5))
ax = sns.barplot(x=popularity['popularity'].head(10),
                y=popularity['original_title'].head(10), 
                data=popularity)
plt.title("most Popular by votes", weight='bold')
plt.xlabel('Score of popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig("best_popular_movie.png")
```

## Recommendation based on scaled weight average and popularity score(Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler()
movie_scaled = scaling.fit_transform(movies_cleaned[['Weighted_average', 'popularity']])
movies_normalized = pd.DataFrame(movie_scaled, columns=['Weighted_average', 'popularity'])
movies_normalized.head()
```

```python
movies_cleaned[['normalized_weight_average','normalized_popularity']] = movies_normalized
```

```python
movies_cleaned.head()
```

```python
movies_cleaned['score'] = movies_cleaned['normalized_weight_average'] * 0.5 + movies_cleaned['normalized_popularity'] * 0.5
movies_scored = movies_cleaned.sort_values(['score'], ascending=False)
movies_scored[['original_title', 'normalized_weight_average','normalized_popularity','score']].head()
```

```python
scored = movies_cleaned.sort_values('score', ascending=False)

plt.figure(figsize=(16,6))
ax = sns.barplot(x=scored['score'].head(5),
                 y=scored['original_title'].head(5),
                data=scored)

plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie title', weight='bold')

plt.savefig('scored_movies.jpeg')
```
